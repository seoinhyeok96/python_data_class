```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
df=pd.read_csv('/Users/mac/Desktop/식물병원.csv')
#필요한 행만 남기기
df = df[['link', 'writer', 'view', 'text', 'YMD', 'time']]
df.head()
```

```python
#수정된 함수
def get_comment(url):
    browser = webdriver.Chrome()
    browser.get(url)
    browser.implicitly_wait(7)

    try:
        comment_list = browser.find_elements(By.CLASS_NAME, 'itemWrap')
    except:
        browser.quit()
        return pd.DataFrame()

    comments = []
    for comment in comment_list:
        try:
            writer = comment.find_element(By.TAG_NAME, 'strong').text
            text = comment.find_element(By.CLASS_NAME, 'txt._commentContent').text.replace('\n','')
            comments.append(f"{writer} : {text}")
        except:
            pass

    browser.quit()
    return comments
```

```python
#열 50개씩 나눠서 추가하기
df1 = df.iloc[:50]#1~50
df1['comment'] = df1['link'].map(get_comment)
```

```python
df1.to_csv('df1.csv', index=False)
pd.read_csv('df1.csv')
```
