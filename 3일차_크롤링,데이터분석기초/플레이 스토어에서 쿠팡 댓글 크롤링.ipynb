```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from selenium import webdriver
from selenium.webdriver import ActionChains
import time
```

```python
PATH = "/Users/seoinhyeog/Desktop/chromedriver"
browser = webdriver.Chrome(PATH)
```

```python
search_word = input("검색하시고자 하는 앱 이름을 입력하세요.")

url = "https://play.google.com/store/search?q={}&c=apps"

final_url = url.format(search_word)
print(final_url)
browser.get(final_url)

time.sleep(3)
browser.implicitly_wait(5)
app_url = browser.find_element(By.CLASS_NAME, 'Si6A0c').get_attribute('href')
print(app_url)

browser.get(app_url)
browser.find_element(By.XPATH, '//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[1]/div[2]/div/div[1]/c-wiz[4]/section/header/div/div[2]/button').click()
```

```python
browser.find_element(By.CLASS_NAME, 'fysCi').click() # 모달창 클릭

modal = browser.find_element(By.CLASS_NAME, 'fysCi')

import time
for i in range(3):
    browser.execute_script("arguments[0].scrollTo(0, arguments[0].scrollHeight);", modal)
    time.sleep(2)
```

```python
import pandas as pd

containers = browser.find_elements(By.CLASS_NAME, 'RHo1pe')

data_list = []

for container in containers:
    
    user_name = container.find_element(By.CLASS_NAME, 'X5PpBb').text
#     print(user_name)
    
    rating = container.find_element(By.CLASS_NAME, 'iXRFPc').get_attribute('aria-label')[10]#별갯수만 나타내게 인덱싱
#     print(rating)
   
    date = container.find_element(By.CLASS_NAME, 'bp9Aid').text
#     print(date)
    
    content = container.find_element(By.CLASS_NAME, 'h3YV2d').text
#     print(content)
    
    data = {
        "user_name": user_name,
        "rating": rating,
        "date": date,
        "content": content        
    }
    
    data_list.append(data)

df = pd.DataFrame(data_list)
df.to_csv('./playstore_comment.csv', encoding = 'utf-8-sig') # csv 파일로 저장
```
