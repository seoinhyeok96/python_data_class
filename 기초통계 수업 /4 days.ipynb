```python
# 와인 데이터 불러오기
df=pd.read_csv("wine.csv")
```

```python
df
```

```python
# 열의 합
df['ash'].sum()
```

```python
# 열의 합
sum_ash = 0
for val in df['ash']:
    sum_ash += val
print(sum_ash)
```

```python
#열의 평균
df['ash'].mean()
```

```python
#열의 평균
sum_ash = 0
for val in df['ash']:
    sum_ash += val
mean_ash = sum_ash/len(df['ash'])
print(mean_ash)
```

```python
# 열의 분산
df['ash'].var()
```

```python
ss_ash=0
for val in df['ash']:
    ss_ash+=(val-mean_ash)**2
var_ash=ss_ash/(len(df['ash'])-1)
print(var_ash)
```

```python
# 열의 표준편차
df['ash'].std()
```

```python
std_ash = var_ash**0.5
print(std_ash)
```

```python
# 공분산
ash = df['ash']
ci = df['color_intensity']
```

```python
np.cov(ash,ci)#cov 공분산
```

```python
n = len(df['ash'])
mean_ash = ash.mean()
mean_ci = ci.mean()
res= 0
for i in range(0,n):
    res += (ash[i]-mean_ash)*(ci[i]-mean_ci)
covariance = res/(n-1)
print(covariance)
```

```python
((ash- mean_ash).T @ (ci- mean_ci)/(n-1))
```
