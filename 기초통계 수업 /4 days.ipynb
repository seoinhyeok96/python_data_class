```python
df=pd.read_csv("star_classification.csv")
```

```python
df
```

```python
df['class'].unique()
```

```python
sub_df = df[["u","g","r","i","z"]]
sub_df
```

```python
# 고유값 분해
sub_df_t = np.transpose(sub_df)
sub_df_t
```

```python
# 고유값 분해
cov_df =sub_df_t.dot(sub_df)
cov_df
```

```python
e, v =np.linalg.eig(cov_df)
```

```python
# 고유값
e
```

```python
# 고유백터
v
```

```python
# 특이값 분해
U,S,Vt = np.linalg.svd(cov_df)
```

```python
#고유 백터
U
```
