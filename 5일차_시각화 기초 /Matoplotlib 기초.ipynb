```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
plt.plot([1,2,3,4,5,4,3,2,1])
plt.show()
```

```python
## sin 함수 그려보보기
x = np.arange(0,12,0.01)# 0~12까지 0.01 간격
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title("sin graph")
plt.show()
```

## 그리드 설정

```python
plt.plot(x,y)
plt.title("sin graph")
plt.grid()
plt.show()
```
